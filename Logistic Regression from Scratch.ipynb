```python
def sigmoid(z):
    return (1/(1+np.exp(-z)))
```

```python
def theta(w, x):
    return w.T.dot(x)
def loss(h, y):
    return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
```

```python
w0 = 1
w1 = 2
w2 = 1
w = np.array([w0, w1, w2], dtype='float64')
w = w.reshape((w.shape[0], 1))
print(w.shape)
print(w)
print(X.shape)
print(X_b.shape)
```

```python
def dj(x, w, y):
    z=theta(w,x)
    y1=y.reshape(y.shape[0],1)
    e = sigmoid(z).T - y1
    return np.dot(x, (e)) / y.shape[0]
```

```python
def optimizar_LMS(x, y, num_iter, alpha, w):
    
    # se generan los vectores
    it = np.arange(0, num_iter)
    j = np.zeros(num_iter)
    
    # Se optimiza el modelo por el numero de iteraciones
    for i in range(num_iter):
        
        # Calculamos la hip√≥tesis
        preds = sigmoid(theta(w,x))

        # Actualizamos los pesos
        grd =dj(x, w, y)
        w = w - alpha * dj(x, w, y)

        # Guardamos el costo
        j[i] = loss(preds, y)
        x1= iris.data[:, 1:2]
        print("Costo", j)
        
      
    return w, j
```

```python
w, j = optimizar_LMS(X_b.T, y, 10000, 0.01,  w)
print("j", j)
preds = sigmoid(theta(w,X_b.T))
```
